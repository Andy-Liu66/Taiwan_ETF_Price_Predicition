## 過程

```python
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
from talib.abstract import *
%matplotlib inline
```

```python
#輸入資料
ETF_data = pd.read_csv("data.csv", encoding = "Big5") #使用調整後股價資料
ETF_data.columns = ["code", "date", "name", "open", "high", "low", "close", "volume"]
ETF_data.date = ETF_data.date.apply(lambda x : datetime.datetime.strptime("{}".format(x), "%Y%m%d"))
```

```python
#檢視有幾種ETF
len(ETF_data.code.unique())#18種ETF
```

```python
#檢查是否有缺值
pd.concat({ "ETF_data" : ETF_data.isna().sum()}, axis = 1) #完全沒缺值
```

以下使用TA-Lib計算技術分析指標做為模型訓練的特徵值。

```python
input_data = ETF_data[ETF_data.code == 50].copy()

#計算技術分析指標
input_data["EMA"] = EMA(input_data, timeperiod = 5) #計算EMA(5)
input_data["EMA"] = EMA(input_data, timeperiod = 10) #計算EMA(10)
input_data["EMA"] = EMA(input_data, timeperiod = 20) #計算EMA(20)
input_data = pd.concat([input_data, BBANDS(input_data, 20, 2, 2)], axis = 1) #計算BBands
input_data = pd.concat([input_data, STOCH(input_data)], axis = 1) #計算KD值
input_data["Momentum"] = MOM(input_data) #計算Momentum
input_data["RSI"] = RSI(input_data) #計算RSI
input_data["WILLR"] = WILLR(input_data) #計算William指標
input_data.volume = input_data.volume.apply(lambda x : float(str.split(x, " ")[-1].replace(",", ""))) #待會要用到成交量資料，但此資料為str因此要先轉成float
input_data["ADOSC"] = ADOSC(input_data) #計算ADOSC
input_data["ATR"] = ATR(input_data) #計算ATR

input_data.dropna(inplace = True) #去除缺值

#選擇要訓練的特徵
input_data = input_data[['open', 'high', 'low', 'close', 'volume', 'EMA', 'upperband', 'middleband', 'lowerband', 'slowk', 'slowd', 'Momentum', 'RSI', 'WILLR', 'ADOSC', 'ATR']]
input_data.head(5)
```

以下將使用類似時間序列模型中AR(p)的概念創造更多特徵值，亦即自我回歸模型，例如p若為5則代表t期的y將由t-1, t-2...t-5期的y所構成(時間序列中要在加上隨機項，但在此處則不考慮隨機項)，以下將套用此概念而非實際使用此模型，因此特徵值為t-1, t-2...t-5期的技術指標及各期的開高低收與成交量，應變數則為t期的收盤價。在此作業中將使用隨機森林的回歸器來預測收盤價，而漲跌部分則由預測出的價格與前日比較，最後將與實際資料比對以計算分數。

```python
#定義函數看要落後幾期
def lag_operator(data, lag):
    
    final_data = pd.DataFrame([])
    for i in np.arange(1, (lag), 1):
        temp_data = data.shift(i)
        
        #重新設columns名字
        temp_data.columns = pd.DataFrame([input_data.columns.values]).T.loc[:, 0].apply(lambda x : "{}_{}".format(x, i + 1)).values
        
        #合併進來
        final_data = pd.concat([final_data, temp_data], axis = 1)
        
    data.columns = pd.DataFrame([data.columns.values]).T.loc[:, 0].apply(lambda x : "{}_{}".format(x, 1)).values
    final_data = pd.concat([data, final_data], axis = 1)
    final_data.dropna(inplace = True)
    return(final_data)
```

```python
#準備資料集

#準備X
X = lag_operator(input_data, 20) #選擇要落後幾期，此處落後20期相當於考量前1個月的交易資料
last_X = X.iloc[-1, :] #另存最後一筆X資料(4/27)，到時可以用來預測4/30的價格
X = X.iloc[:-1, :] #去除最後一筆X資料，因為資料集中並沒有4/30日之價格資料，因此對於4/27日的自變數而言會有沒ylabel，所以剃除

#準備Y
Y = input_data.close_1.shift(-1)[0: -1] #將Y前移一期
Y = Y[(X.index[0] - Y.index[0]) : ] #選擇相對應於X的index的資料區間(因為X前幾期有去除)

#將X, Y依據0.7, 0.3之比例分為訓練資料集及測試資料集
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test =  train_test_split(X, Y, test_size = 0.3, random_state = 666)
```

以上採用隨機的方式分割測試與訓練資料集，若依據時間(例如：前7成，後3成)切割可能會有瑕疵，因為後面若有(結構性)變動前面可能學不到，且後期的價格與前期差距很大。

```python
%%time
#使用RandomizedSearchCV在給定所有的模型參數空間中，透過隨機抽取參數的方式建立模型，最後選擇mse最佳者做為最終採用模型
from sklearn.model_selection import RandomizedSearchCV

#設定模型參數區間
rfr_param_grid = {"n_estimators" : np.arange(10, 31, 1),
                  "criterion" : ["mse"],
                  "max_features" : ["auto", "sqrt", "log2"],
                  "max_depth" : np.arange(1, 31, 1),
                  "min_samples_leaf" : np.arange(1, 11, 1)
                 }

#同樣選擇隨機森林演算法，此處採用回歸模型
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor(random_state = 666) #rfr == random forest regressor

#將模型導入GridSearchCV，同時考量交叉檢驗
rfr_GridSearchCV = RandomizedSearchCV(estimator = rfr, 
                                      param_distributions = rfr_param_grid, 
                                      scoring = "neg_mean_squared_error",
                                      cv = 10,
                                      n_jobs = 3,
                                      n_iter = 30) #隨機抽取30組參數

#訓練GridSearchCV模型
rfr_GridSearchCV.fit(x_train, y_train)

#選擇GridSearchCV模型中表現最佳者做為最終採用模型
rfr = rfr_GridSearchCV.best_estimator_
rfr.fit(x_train, y_train)
```

```python
#以下評估此預測結果的表現，之後最終結果中並不會有此部分

#訓練資料集預測出之結果
y_predict_train = rfr.predict(x_train)

#測試資料集預測出之結果
y_predict_test = rfr.predict(x_test)
```

```python
#整理預測結果並畫圖

#合併訓練資料及與測試資料集的預測結果
predict = pd.DataFrame({"Index" : y_test.index.append(y_train.index), "Values" : np.concatenate([y_predict_test, y_predict_train])})

#由於是隨機取值因此index是亂數，要重新指定然後依順序排好才能畫圖
predict.index = predict.Index
predict.sort_index(inplace = True)

#畫圖，可以發現圖形看起來預測得還不錯，但因為是圖形scale的問題，以下將另外繪圖看各天預測情形
y_predict_train = rfr.predict(x_train)
y_predict_test = rfr.predict(x_test)
plt.figure(figsize = (20, 10))
plt.plot(predict.Index, predict.Values,  label = "predict")
# plt.scatter(y_test.index, y_predict_test, label = "predict", c = "green")
# plt.scatter(y_train.index, y_predict_train, label = "predict")
plt.plot(Y.index, Y.values, label = "real")
plt.legend(loc = 2, fontsize = "x-large");
```

```python
#以下分別檢視預測資料集與測試資料集的誤差狀況，圖中柱狀圖是由預測值減去實際值，若bar大於0則代表預測高估，反之亦然
#可以發現訓練資料集的表現較好(y軸預測錯誤的range較小)，因此存有overfitting的問題，但以下仍採用此模型
opacity = 0.6 
fig, ax = plt.subplots(2, 1, figsize = (20, 15))

ax1 = ax[0]
ax1.bar(np.arange(1, len(y_predict_train) + 1 , 1), (y_predict_train - y_train), alpha = opacity, color = "blue")
ax1.title.set_text(("Train_set"))

ax2 = ax[1]
ax2.bar(np.arange(1, len(y_predict_test) + 1 , 1), (y_predict_test - y_test), alpha = opacity, color = "red")
ax2.title.set_text(("Test_set"));
```

```python
#預測4/30日之價格，使用上述保留的4/27資料(4/27沒有被拿去訓練，已被剔除)
rfr.predict(last_X.values.reshape(1, x_train.shape[1])) #4/30實際股價為80，差距不會太遠
```

## 最終結果
以下開始將上述過程(排除繪圖)透過for loop套用至各ETF，並回傳預測值。

```python
%%time
ETF_code = ETF_data.code.unique()
result = pd.DataFrame([])
for i in range(len(ETF_code)):
    #選擇當前股票
    input_data = ETF_data[ETF_data.code == ETF_code[i]].copy()

    
    #計算技術分析指標
    input_data["EMA"] = EMA(input_data, timeperiod = 5) #計算EMA(5)
    input_data["EMA"] = EMA(input_data, timeperiod = 10) #計算EMA(10)
    input_data["EMA"] = EMA(input_data, timeperiod = 20) #計算EMA(20)
    input_data = pd.concat([input_data, BBANDS(input_data, 20, 2, 2)], axis = 1) #計算BBands
    input_data = pd.concat([input_data, STOCH(input_data)], axis = 1) #計算KD值
    input_data["Momentum"] = MOM(input_data) #計算Momentum
    input_data["RSI"] = RSI(input_data) #計算RSI
    input_data["WILLR"] = WILLR(input_data) #計算William指標
    input_data.volume = input_data.volume.apply(lambda x : float(str.split(x, " ")[-1].replace(",", ""))) #待會要用到成交量資料，但此資料為str因此要先轉成float
    input_data["ADOSC"] = ADOSC(input_data) #計算ADOSC
    input_data["ATR"] = ATR(input_data) #計算ATR
    input_data.dropna(inplace = True)
    #選擇要訓練的特徵
    input_data = input_data[['open', 'high', 'low', 'close', 'volume', 'EMA', 'upperband', 'middleband', 'lowerband', 'slowk', 'slowd', 'Momentum', 'RSI', 'WILLR', 'ADOSC', 'ATR']]
    
    
    
    #準備資料集
    #準備X
    X = lag_operator(input_data, 20) #選擇要落後幾期
    last_X = X.iloc[-1, :] #另存最後一筆X資料(4/27)，到時可以用來預測4/30的價格
    X = X.iloc[:-1, :] #去除最後一筆X資料，因為資料集中並沒有4/30日之價格資料，因此對於4/27日的自變數而言會有沒ylabel，所以剃除
    #準備Y
    Y = input_data.close_1.shift(-1)[0: -1] #將Y前移一期
    Y = Y[(X.index[0] - Y.index[0]) : ] #選擇相對應於X的index的資料區間(因為X前幾期有去除)
    #將X, Y依據0.7, 0.3之比例分為訓練資料集及測試資料集
    x_train, x_test, y_train, y_test =  train_test_split(X, Y, test_size = 0.3, random_state = 666)
    
    
    
    #使用RandomizedSearchCV在給定所有的模型參數空間中，透過隨機抽取參數的方式建立模型，最後選擇mse最佳者做為最終採用模型
    #設定模型參數區間
    rfr_param_grid = {"n_estimators" : np.arange(10, 31, 1),
                  "criterion" : ["mse"],
                  "max_features" : ["auto", "sqrt", "log2"],
                  "max_depth" : np.arange(1, 31, 1),
                  "min_samples_leaf" : np.arange(1, 11, 1)
                 }
    #同樣選擇隨機森林演算法，此處採用回歸模型
    rfr = RandomForestRegressor(random_state = 666) #rfr == random forest regressor
    #將模型導入GridSearchCV
    rfr_GridSearchCV = RandomizedSearchCV(estimator = rfr, 
                                      param_distributions = rfr_param_grid, 
                                      scoring = "neg_mean_squared_error",
                                      cv = 10,
                                      n_jobs = 3,
                                      n_iter = 30) #隨機抽取30組參數，此處並未指定random_state，因此每次結果可能有所不同
    #訓練GridSearchCV模型
    rfr_GridSearchCV.fit(x_train, y_train)
    #選擇GridSearchCV模型中表現最佳者做為最終採用模型
    rfr = rfr_GridSearchCV.best_estimator_
    rfr.fit(x_train, y_train)
    
    
    #儲存最終預測結果
    temp_result = pd.DataFrame([ETF_code[i], rfr.predict(last_X.values.reshape(1, x_train.shape[1]))[0]]).T
    result = pd.concat([result, temp_result])
    
    print("ETF-{} finished, {} remain...".format(ETF_code[i], 18 - (i + 1)))
```

```python
#接著看預測結果及分數，分數計算方式依據官網公告規則

#預測值
result.columns = ["Code", "predict_20180430"]
result.Code = result.Code.apply(lambda x : int(x))

#4/30日實際股價，從TEJ下載，使用資料為調整後股價資料
correct_answer = pd.read_csv("etf_430_selected.csv")
correct_answer = correct_answer.iloc[:, [3, 4]]
correct_answer.columns = ["close_20180430", "Code"]

#4/27日實際股價，從一開始資料集獲取
actual_data = ETF_data[ETF_data.date == datetime.datetime.strptime("20180427", "%Y%m%d")][["code", "close"]]
actual_data.columns = ["Code", "close_20180427"]

#整理合併上述表格
result = result.merge(correct_answer, on = "Code").merge(actual_data, on = "Code")

#計算漲跌
result["predict_trend"] = (result["predict_20180430"] - result["close_20180427"])
result["actual_trend"] = (result["close_20180430"] - result["close_20180427"])

#將漲跌轉換為0,1分別代表跌或漲
result["predict_trend"] = result["predict_trend"].apply(lambda x : 0 if x <=0 else 1)
result["actual_trend"] = result["actual_trend"].apply(lambda x : 0 if x <=0 else 1)

#計算成績
#來自漲跌的成績
score_trend = (result.predict_trend == result.actual_trend).sum() * 0.5
#來自預測價格的成績
score_close = (((result.predict_20180430 - abs(result.predict_20180430 - result.close_20180430)) / result.close_20180430) * 0.5).sum()

print("最終成績為:{}分 (滿分為18分)".format(round(score_trend + score_close, 4)))
```
